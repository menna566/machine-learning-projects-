```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt  
import seaborn as sns
from plotnine import *
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename)) 
        f_name=os.path.join(dirname,filename)
```

```python
data=pd.read_csv(f_name)
data.head(40)
```

```python
#to help me to get the percentage of how many people left the job and how many pof them stayed
data['Attrition'].value_counts()
```

```python
data["Attrition"].replace("Yes",1,inplace=True)
data["Attrition"].replace("No",0,inplace=True)
```

```python
'''the percentage tells me that 16,12% of employees left the job at IBM 
and 83.88 still working there'''
plt.figure(figsize= (10, 6))
data_pie  = [237 , 1233]
labels = ["Yes", "No"]
explode = [0.1, 0]
plt.pie(data_pie ,labels= labels , explode = explode , autopct="%1.2f%%", shadow= True, colors= ['#256D85', '#3BACB6'])
plt.legend()
plt.show()
```

```python
data[(data['Gender']=="Male") & (data['Attrition']==1)]
```

```python
#the percntage of the Females who left the job less than the percentage of males 
sns.set_theme(style="darkgrid")
ax = sns.countplot(x="Attrition", hue='Gender', data=data)
```

```python
#most of the employees who didnt leave have good montlyincome but those who left 5% of them just has high income the others were around the avarage
sns.catplot(x="Attrition", y="MonthlyIncome", data=data)
```

```python
sns.lineplot(data=data, x="DailyRate", y="MonthlyIncome",hue="Attrition")
```

```python
#if we connected the dots we will find a relation between the dailyrate and decreasing the monthly income 
#as they work very hard so their daily rate is high but still thier salary is low 
sns.catplot(x="Attrition", y="DailyRate", kind="box", data=data)
sns.catplot(x="Attrition", y="DailyRate", data=data)
```

```python
#5% of the people that left took overtime
#maybe we will try to give more people overtime so less people leave
ax = sns.countplot(x="Attrition", data=data,hue="OverTime",palette='spring')
for container in ax.containers:
     ax.bar_label(container)
```

```python
#the more you stay at the company less chance you have to be promoted may mean that people are not really growing within the company
ax = sns.countplot(x="YearsSinceLastPromotion", data=data,hue="Attrition", palette='winter')
```

```python
#15% of people who leave travel rerely maybe we should give them more work trips to raise their salary 
ax = sns.countplot(x="Attrition", data=data,hue="BusinessTravel",palette='spring')
```

```python
'''high correlation between these two features
apperently the longer you are in the company less chance you have to be promoted
may mean that people are not really growing within the company'''

ax = sns.stripplot(x="YearsSinceLastPromotion", y="YearsAtCompany", data=data,palette="Set2")
```

```python
#higher correlation between years since last promotion and years at company for those who don't work overtime
ax =  sns.stripplot(x="YearsSinceLastPromotion", y="YearsAtCompany", hue="OverTime",
                   data=data, palette="Set2", size=20, marker="D",
                   edgecolor="gray", alpha=.25)
'''g = sns.catplot(x="YearsSinceLastPromotion", y="YearsAtCompany",
                hue="OverTime",
                data=data, kind="strip",
                height=4, aspect=.7); '''
```
